```
import requests
from bs4 import BeautifulSoup

from tqdm.notebook import tqdm
```

```
# 이후부터는 네이버웹툰 섬네일 가져온 거랑 똑같다.
# img src 링크들 수집  
url = 'https://www.google.com/search?q=%EA%B0%95%EC%95%84%EC%A7%80&hl=ko&gbv=1&sxsrf=ALiCzsbdPqbNnCtmNkHYM00zcchWdsx0PQ:1657845228074&source=lnms&tbm=isch&sa=X&ved=2ahUKEwjeu6Xf0vn4AhWlRWwGHaTtC0YQ_AUoAXoECAMQAw&biw=1280&bih=577&dpr=1.5&sfr=gws&sei=ArbQYtaOFYeM0gSC0azoBA'

resp = requests.get(url)
resp
```

```
soup = BeautifulSoup(resp.content, 'lxml')
img_tags = soup.select('img.yWs4tf')
len(img_tags)

for idx, img_tag in enumerate(img_tags):
    img_src = img_tag['src']
    resp_img = requests.get(img_src)

    img = resp_img.content

    with open('강아지_{:02}.jpg'.format(idx+1), 'wb') as f:
        f.write(img)
```

```
# 여러페이지 

for i in tqdm(range(10), desc='페이지'):
    # i = 0, 1, 2, 3
    # start = 0, 20, 40, 60
    start = 20*i
    url = 'https://www.google.com/search?q=%EA%B0%95%EC%95%84%EC%A7%80&hl=ko&gbv=1&biw=1280&bih=577&tbm=isch&sxsrf=ALiCzsbwaZdxp_qpN2ONgFlxSwo1KGciTw:1657845891787&ei=g7jQYv28L_e12roPuoa5sAU&start={}&sa=N'.format(start)
    resp = requests.get(url)

    soup = BeautifulSoup(resp.content, 'lxml')
    img_tags = soup.select('img.yWs4tf')
    len(img_tags)

    for idx, img_tag in tqdm(enumerate(img_tags), desc='이미지/페이지'):
        img_src = img_tag['src']
        resp_img = requests.get(img_src)

        img = resp_img.content

        with open('강아지_{:03}.jpg'.format(start+idx+1), 'wb') as f:
            f.write(img)
```
