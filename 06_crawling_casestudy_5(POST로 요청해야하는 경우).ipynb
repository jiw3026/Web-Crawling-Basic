# POST로 요청해야하는 경우
- request method : CRUD
    1. Create : POST  
    2. Read : GET
    3. Update  
    4. Delete 

올리브영 상품 정보  
https://www.oliveyoung.co.kr/store/goods/getGoodsDetail.do?goodsNo=A000000117541&dispCatNo=100000100010008&trackingCd=Cat100000100010008_MID

- request url 을 찾았을 때, 바로 아래에 request method가 GET인지, POST인지 확인하기
- POST라면 payload 탭열기
- form data 그대로 복사해서 딕셔너리로 만들기
- requests.post(url, data=data) (만약 안되면, data대신 json=data)

```python
import requests
from bs4 import BeautifulSoup
```

```python
url = 'https://www.oliveyoung.co.kr/store/goods/getGoodsArtcAjax.do'

data = {
    'goodsNo': 'A000000169069',
    'itemNo': '001',
    'pkgGoodsYn': 'N'
}

headers = {
    'referer': 'https://www.oliveyoung.co.kr/store/goods/getGoodsDetail.do?goodsNo=A000000169069&dispCatNo=10000010008&trackingCd=Cat10000010008_Prodrank',
    'user-agent': 'Mozilla/5.0 (Windows NT 10.0; Win64; x64) AppleWebKit/537.36 (KHTML, like Gecko) Chrome/103.0.0.0 Safari/537.36'
}


resp = requests.post(url, data=data, headers=headers)
print(resp.text)
```

```python
soup = BeautifulSoup(resp.text, 'lxml')
soup.text
```
