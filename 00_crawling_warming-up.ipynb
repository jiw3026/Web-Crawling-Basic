```python
import requests
from bs4 import BeautifulSoup
```

```python
url = 'https://search.naver.com/search.naver?where=news&ie=utf8&sm=nws_hty&query=%EC%BD%94%EB%A1%9C%EB%82%98+%ED%99%95%EC%A7%84%EC%9E%90'

resp = requests.get(url)
resp
```

requests.get(url) return 값 정리
- 응답코드 (status_code) 2xx > ok, 4xx > no (1xx, 2xx, 3xx, 4xx, 5xx)
- html (content, text)

```python
html = resp.content
```

html의 타입을 beautifulSoup형으로 변형해줘야 한다.

```python
soup = BeautifulSoup(html, 'lxml')
# 이제, select 함수를 이용해서
# 전체 html에서 원하는 태그를 가져올 수 있다.
```

```python
# 전체.select(원하는태그)
a_tags = soup.select('a.news_tit') # class가 news_tit인 a태그
len(a_tags)
```

태그 중에서 제목을 뽑아내는 게 핵심
- 태그 속성은 딕셔너리라고 생각하자.

```python
a_tags[0]['title'], a_tags[0]['href']
```

```python
titles, urls = [], []

for a_tag in a_tags:
    titles.append(a_tag['title'])
    urls.append(a_tag['href'])
```

```python
df_news = {
    '제목' : titles,
    '링크' : urls
}
```

```python
import requests
from bs4 import BeautifulSoup
```

```python
url = 'https://search.naver.com/search.naver?where=news&ie=utf8&sm=nws_hty&query=%EC%82%BC%EC%84%B1%EC%A0%84%EC%9E%90'

resp = requests.get(url)
resp
```

```python
soup = BeautifulSoup(resp.content, 'lxml')
tit_tags = soup.select('a.new_tit')
key_tags = soup.select('a.keyword div') # 스페이스 = 상위 - 하위 연결자(~안에 있는, ~아래에 있는)
len(tit_tags), len(key_tags)
```

```python
tits, keys = [], []

for tit_tag, key_tag in zip(tit_tags, key_tags):
    # print(tit_tag['title'])
    # print(key_tag.text)

    tits.append(tit_tag['title'])
    keys.append(key_tag.text)

tits, keys
```

```python
samsung_info = {
    '기사제목' : tits,
    '연관검색어' : keys
}

samsung_info
```

```python
import pandas as pd
```

```python
df_samsung = pd.DataFrame(samsung_info)
df_samsung
```
